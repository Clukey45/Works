Background: This project utilizes python along with its librarys to perform web scraping; with the goal of extracting, organizing, and analyzing event information from the Ocean Beach MainStreet Association website.

Librarys used: requests, BeautifulSoup, datetime, pandas 



```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd 
from datetime import datetime, timedelta
```

```python
def scrape_ob_calendar():
    url = 'https://oceanbeachsandiego.com/calendar/month'
    months = ['2025-'+ f"{i:02d}" for i in range(1,13)]
    urllist = [ url+"/"+month for month in months]
    df = pd.DataFrame()
    for url in urllist:
        response = requests.get(url)
        soup = bs(response.text, 'html.parser')
        tbody = soup.find('tbody')
        for tr in tbody.find_all('tr', class_=['single-day','multi-day']):
            classy = tr.get('class')
            if classy[0] == 'single-day':
                for td in tr.find_all('td'):
                    if td.find('a') and td.find('span'):
                        date = str(td.get('data-date'))
                        date = datetime.strptime(date, "%Y-%m-%d")
                        month = date.strftime("%B")
                        day = td.get('headers') 
                        start = td.find('span', class_='date-display-start')
                        end = td.find('span', class_='date-display-end')
                        if start and end:
                            start = start.get_text(strip=True)
                            end = end.get_text(strip=True)
                        else:
                            single = td.find('span', class_='date-display-single')
                            start = single.get_text(strip=True) if single else None
                            end = None
                        atag = td.find_all('a')
                        for a in atag:
                            d = {}
                            event = a.get_text(strip=True)
                            d['class'] = classy
                            d['day'] = day
                            d['date'] = date
                            d['month'] = month
                            d['event'] = event
                            d['start'] = start
                            d['end'] = end
                            temp = pd.DataFrame.from_dict(d)
                            df = pd.concat([df,temp], ignore_index = True)
            elif classy[0] == 'multi-day':
                for td in tr.find_all('td'):
                    if td.find('a') and td.find('span'):
                        span = int(td.get('colspan'))
                        sd = str(td.get('data-date'))
                        sd = datetime.strptime(sd, "%Y-%m-%d")
                        for i in range(span):
                            date = sd + timedelta(days = i)
                            day = td.get('headers')
                            month = date.strftime("%B")
                            start = td.find('span', class_='date-display-start')
                            end = td.find('span', class_='date-display-end')
                            if start and end:
                                start = start.get_text(strip=True)
                                end = end.get_text(strip=True)
                            else:
                                single = td.find('span', class_='date-display-single')
                                start = single.get_text(strip=True) if single else None
                                end = None
                            atag = td.find_all('a')
                            for a in atag:
                                d = {}
                                event = a.get_text(strip=True)
                                d['class'] = classy
                                d['day'] = day
                                d['date'] = date
                                d['month'] = month
                                d['event'] = event
                                d['start'] = start
                                d['end'] = end
                                temp = pd.DataFrame.from_dict(d)
                                df = pd.concat([df,temp], ignore_index = True)
    return df
```
